```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
import numpy as np
import matplotlib.pyplot as plt
```

```python
def price(ticker, period='1d',columns=['Close']):
    '''
    Returns a DataFrame of prices for ticker from Yahoo Finance API 
    '''
    obj = yf.Ticker(ticker)
    return obj.history(period=period)[columns]
```

```python
def prices(tickers, period='1d',columns=['Close']):
    '''
    Returns a DataFrame of prices for a list of tickers from Yahoo Finance API 
    '''
    prices = pd.DataFrame()
    column_names = [col + ' ' + ticker for ticker in tickers for col in columns]
    __prices = [price(ticker, period, columns) for ticker in tickers]
    prices = pd.concat(__prices, axis=1)
    prices.columns=column_names
    return prices
```

```python
def normalize(col):
    return col/col[0]
```

```python
stonks = ['GOOG', 'MSFT', 'TSLA']
etfs = ['EUNL', 'EIMU', 'EXSI']
columns=['Open', 'Close']
```

```python
stonk_data = prices(stonks, period='1y', columns=columns)
```

```python
stonk_data.apply(normalize).plot(logy=True)
```

```python
stonk_data.apply(normalize).plot(logy=False)
```
